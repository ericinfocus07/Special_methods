```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)


class Student(Person):

    def __init__(self, name, age, student_id, enroll_course):
        super().__init__(name, age)
        self.student_id = student_id
        self.enroll_course = enroll_course


    def display_info(self):
        super().display_info()
        print("Student ID:", self.student_id)

    def enroll_course_method(self):
        print(self.name, " has a student ID and can enroll in", self.enroll_course)


class Lecturer(Person):

    def __init__(self, name, age, employee_id, teach_course):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.teach_course = teach_course

    def teach_course_method(self):
        print(self.name, "has an employee ID and can teach", self.teach_course)



student1 = Student("Eric", 19, "FOE.41.006.091.25", "Object-Oriented Programming")
lecturer1 = Lecturer("Mr Cobbinah", 45, "ST001", "Object-Oriented Programming")

print("------ STUDENT ------")
student1.display_info()
student1.enroll_course_method()

print()

print("------ LECTURER ------")
lecturer1.display_info()
lecturer1.teach_course_method()
```
